```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('heart.csv')
df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

str_col = ['Sex', 'ChestPainType', 'RestingECG', 'MaxHR', 'ExerciseAngina', 'ST_Slope']

for i in str_col:
  le = le.fit(df[i])
  df[i] = le.transform(df[i])
df
```

```python
target = df.pop('HeartDisease')
target
```

```python
dataset = tf.data.Dataset.from_tensor_slices((df.values, target.values))
dataset_p = tf.data.Dataset.from_tensor_slices((df['ChestPainType'].values, target.values))
```

```python
ChestPainType_input = keras.Input(shape=(1,), name='ChestPainType')
embedding = keras.layers.Embedding(df['ChestPainType'].max()+1, 4)(ChestPainType_input)
model = keras.Model(inputs= ChestPainType_input, outputs = embedding)

model.predict(df['ChestPainType'])
```
